**#29. На сколько процентов ежемесячно менялось количество заказов в компании Northwind с 1 апреля по 1 декабря 2017 года? Отобразите таблицу со следующими полями:**\
**- номер месяца;**\
**- количество заказов в месяц;**\
**- процент, который показывает, насколько изменилось количество заказов в текущем месяце по сравнению с предыдущим.**\
**Если заказов стало меньше, значение процента должно быть отрицательным, если больше — положительным. Округлите значение процента до двух знаков после запятой. Отсортируйте таблицу по возрастанию значения месяца.**

```python
query_29 = """
WITH
pre_month AS (
    SELECT  CAST(strftime('%m', OrderDate) + 1 AS int) month_number,
            CAST(COUNT(OrderID) AS REAL) orders_by_pre_month
    FROM    Orders
    WHERE   CAST(strftime('%Y', OrderDate) AS int) = 2017
    GROUP BY CAST(strftime('%m', OrderDate) AS int)
),
current_mounth AS (
    SELECT  CAST(strftime('%m', OrderDate) AS int) month_number,
             CAST(COUNT(OrderID) AS REAL) orders_by_month
    FROM    Orders
    WHERE   CAST(strftime('%Y', OrderDate) AS int) = 2017
    GROUP BY CAST(strftime('%m', OrderDate) AS int)
)


SELECT  cur.month_number,
        cur.orders_by_month,
        ROUND((orders_by_month - pre.orders_by_pre_month) / pre.orders_by_pre_month * 100, 2) percentage
FROM    current_mounth cur
INNER JOIN   pre_month AS pre ON cur.month_number=pre.month_number
WHERE cur.month_number BETWEEN 4 AND 12
ORDER BY cur.month_number;
"""
result_29 = pd.read_sql_query(query_29, conn)
```

```python
result_29
```

**#30. На сколько процентов ежегодно менялось количество заказов в Northwind с 2016 по 2018 годы. Отобразите таблицу со следующими полями:**\
**- Число года.**\
**- Количество заказов за год.**\
**- Процент, округлённый до целого числа, который показывает, насколько изменилось количество заказов в текущем году по сравнению с предыдущим.**\
**Для 2016 года выведите значение NULL.**

```python
query_30 = """
WITH
pre_year AS (
    SELECT  CAST(strftime('%Y', OrderDate) + 1 AS int) year_number,
            CAST(COUNT(OrderID) AS REAL) orders_by_pre_year
    FROM    orders
    WHERE   CAST(strftime('%Y', OrderDate) AS int) BETWEEN 2016 AND 2018
    GROUP BY CAST(strftime('%Y', OrderDate) AS int)
),
current_year AS (
    SELECT  CAST(strftime('%Y', OrderDate) AS int) year_number,
            CAST(COUNT(OrderID) AS REAL) orders_by_year
    FROM    orders
    WHERE   CAST(strftime('%Y', OrderDate) AS int) BETWEEN 2016 AND 2018
    GROUP BY CAST(strftime('%Y', OrderDate) AS int)
)


SELECT  cur.year_number,
        orders_by_year,
        ROUND((orders_by_year - pre.orders_by_pre_year) / pre.orders_by_pre_year * 100, 2) percentage
FROM    current_year cur
FULL OUTER JOIN   pre_year AS pre ON cur.year_number=pre.year_number
WHERE cur.year_number BETWEEN 2016 AND 2018
ORDER BY cur.year_number;
"""
result_30 = pd.read_sql_query(query_30, conn)
```

```python
result_30
```

**#31. Рассчитайте аналог Retention Rate по неделям для компаний-заказчиков. Объедините компании в когорты по неделе их первого заказа (поле OrderDate). Возвращение определяйте по наличию заказа в течение текущей недели. Значение Retention Rate округлите до двух знаков после запятой.**

```python
query_31 = """
WITH
-- Определяем дату первого заказа (день рождения когорты) для каждого клиента:
cohorts AS (
    SELECT 
        CustomerID,
        date(MIN(OrderDate), 'weekday 1', '-7 days') AS cohort_week
    FROM orders
    GROUP BY CustomerID
),
-- Группируем все заказы по неделям:
activities AS (
    SELECT 
        CustomerID,
        date(OrderDate, 'weekday 1', '-7 days') AS order_week
    FROM orders
    GROUP BY CustomerID, order_week
),
-- Считаем количество клиентов в каждой когорте:
cohort_sizes AS (
    SELECT 
        cohort_week,
        COUNT(CustomerID) AS cohort_size
    FROM cohorts
    GROUP BY cohort_week
),
-- Соединяем данные, чтобы увидеть активность клиентов по неделям относительно их первой покупки:
retention_table AS (
    SELECT 
        c.cohort_week,
        a.order_week,
        COUNT(DISTINCT a.CustomerID) AS active_customers
    FROM cohorts c
    JOIN activities a ON c.CustomerID = a.CustomerID
    GROUP BY c.cohort_week, a.order_week
)

-- Финальный расчет Retention Rate:
SELECT 
    r.cohort_week,
    r.order_week,
    r.active_customers,
    s.cohort_size,
    ROUND(CAST(r.active_customers AS REAL) / s.cohort_size * 100, 2) AS retention_rate
FROM retention_table r
JOIN cohort_sizes s ON r.cohort_week = s.cohort_week
ORDER BY r.cohort_week, r.order_week;
"""
result_31 = pd.read_sql_query(query_31, conn)
```

```python
result_31
```

**#34. Для каждой компании, которая оформила хотя бы два заказа, выведите:**\
**- дату оформления второго заказа (поле OrderDate), округлённую до месяца**\
**- идентификатор компании, оформившей заказ (поле CustomerID)**\
**Строки отсортируйте по значению в поле с идентификаторами в лексикографическом порядке по убыванию.**

```python
query_34 = """
WITH RankedOrders AS (
    SELECT 
        CustomerID,
        OrderDate,
        -- Нумеруем заказы для каждого клиента по дате оформления
        ROW_NUMBER() OVER (PARTITION BY CustomerID ORDER BY OrderDate, OrderID) as rn
    FROM orders
)


SELECT 
    -- Округляем дату до первого числа месяца
    date(OrderDate, 'start of month') order_date,
    CustomerID
FROM RankedOrders
WHERE rn = 2
ORDER BY CustomerID DESC;
"""
result_34 = pd.read_sql_query(query_34, conn)
```

**#35. Для каждого месяца с июля 2016 года по май 2018 года посчитайте конверсию в процентах. Найдите количество уникальных компаний-заказчиков в текущем месяце. Разделите его на общее количество компаний-заказчиков Northwind, которые оформили хотя бы один заказ за всё предыдущее время, включая текущий месяц. Под уникальностью компании в этой задаче подразумевается отсутствие повторов в выборке.**
**В итоговой таблице должны быть следующие поля:**\
**- дата первого числа текущего месяца;**\
**- количество компаний-заказчиков в текущий месяц;**\
**- общее количество компаний-заказчиков за всё предыдущее время, включая текущий месяц;**\
**- отношение количества покупателей за текущий месяц к общему количеству покупателей.**\
**При делении не забудьте привести значения к вещественному типу real, тогда после деления вы получите вещественное, а не целое число. Не забудьте умножить результат на 100 и округлить значение до двух знаков после запятой.**

```python
query_35 = """
-- Считаем количество уникальных клиентов для каждого месяца:
WITH MonthlyStats AS (
    SELECT 
        date(OrderDate, 'start of month') month_date,
        COUNT(DISTINCT CustomerID) monthly_customers
    FROM orders
    WHERE OrderDate BETWEEN '2016-07-01' AND '2018-05-31'
    GROUP BY 1
),
-- Считаем количество клиентов, сделавших хотя бы один заказ до текущего месяца включительно:
CumulativeStats AS (
    SELECT 
        month_date,
        monthly_customers,
        (SELECT COUNT(DISTINCT CustomerID) 
         FROM orders o
         WHERE  date(OrderDate, 'start of month') <= m.month_date) AS total_customers
    FROM MonthlyStats m
)
-- Вычисляем конверсию с приведением типов и округлением:
SELECT 
    month_date,
    monthly_customers,
    total_customers,
    ROUND(
        CAST(monthly_customers AS REAL) / total_customers * 100, 
        2
    ) AS conversion_rate
FROM CumulativeStats
ORDER BY month_date;
"""
result_35 = pd.read_sql_query(query_35, conn)
```
